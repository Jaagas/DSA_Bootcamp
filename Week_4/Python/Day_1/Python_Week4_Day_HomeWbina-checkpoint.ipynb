```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url= 'http://bina.az'
response = requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

## LINKS 

```python
for i in soup.find_all('a',class_='item_link'):
    
```

```python
links=[]
for i in soup.find_all('a',class_='item_link'):
  
    links.append("https://bina.az/"+i.get('href'))
links
```

## LOCATION

```python
locations=[]
for i in soup.find_all('div',class_='location'):
    locations.append(i.text)
locations
```

## DATA and TIME

```python
data_time = []
for i in soup.find_all('div',class_='city_when'):
    
    data_time.append(i.text.split(",")[1])
data_time   
```

## PRICE

```python
price=[]
for i in soup.find_all('div',class_='price'):
    price.append(i.text)
price
```

```python
bina=list(zip(locations,data_time,price,links))
```

```python
bina_colums = ['location',"data","price","links"]
pd.DataFrame(bina,columns=bina_colums)
```
