## 16. Show footballer has maximum Shot Power.

```python

team = tr_df.groupby('Club')

greatest_club_players_df = pd.DataFrame(columns=tr_df.columns)
for clb, data in team:
    best_player = data.nlargest(1, 'ShotPower')
    greatest_club_players_df = greatest_club_players_df.append(best_player)
    
greatest_club_players_df.sort_values('ShotPower', ascending=False)



```
